Data Cleaning In Pandas

```python
import pandas as pd
```

```python
df1 = pd.read_excel(r"C:\Users\kedugga\Desktop\Data Analyst\Python\Pandas\Project 1\Customer Call List.xlsx")
df1
```

```python
# Droping duplicates
df1= df1.drop_duplicates()
df1
```

```python
#Dropping a column
df1 = df1.drop(columns= 'Not_Useful_Column')
df1
```

```python
#Cleaning Last_name Column
df1['Last_Name']=df1['Last_Name'].str.strip("123./")
df1
```

```python
# To remove non-alphanumeric characters from a Phone_Number column
df1["Phone_Number"]=df1["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
df1
```

```python
#Formatting the column by using lamda
df1["Phone_Number"]= df1["Phone_Number"].apply(lambda x: str(x))
df1
#df["Phone_Number"].apply(lambda x: x[0:3]+ '_' + x[3:6]+ '_' + x[6:10])
```

```python
df1["Phone_Number"]= df1["Phone_Number"].apply(lambda x: x[0:3]+ '-' + x[3:6]+ '-' + x[6:10])
df1
```

```python
# Removing nan, Na from the Phone Number Column
df1["Phone_Number"]= df1["Phone_Number"].str.replace('nan--','')
df1["Phone_Number"]= df1["Phone_Number"].str.replace('Na--','')
df1
```

```python
#Splitting the Address Column
df1[["Street Address","State","ZipCode"]]= df1["Address"].str.split(',', n=2, expand=True)
df1
```

```python
df1["Paying Customer"]=df1["Paying Customer"].str.replace('Yes','Y')
df1["Paying Customer"]=df1["Paying Customer"].str.replace('No','N')
df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('Yes','Y')
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('No','N')
df1
```

```python
df1.replace('NaN','')
```

```python
df1.fillna('')
```

```python
#Filling the None with space
df1= df1.fillna('')
df1
```

```python
for x in df1.index:
    if df1.loc[x,"Do_Not_Contact"] == 'Y':
        df1.drop(x,inplace=True)

df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].replace('','N')
df1
```

```python
for x in df1.index:
    if df1.loc[x,"Phone_Number"] == '':
        df1.drop(x,inplace=True)

df1
```

```python
df1= df1.reset_index(drop=True)
df1
```

```python
df1.drop("ZipCode",axis=1,inplace=True)
df1
```

```python
df1.drop("State",axis=1,inplace=True)
df1
```
